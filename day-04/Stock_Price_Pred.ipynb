**Apple Stock Price Prediction**

Import Required Libraries

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
```

Read the Dataset

```python
data = pd.read_csv('AAPL.csv')
```

```python
data.head()
```

Visualizing the dataset

```python
figure = go.Figure(data=[go.Candlestick(x=data['Date'],
                                        open=data['Open'], high=data['High'],
                                        low=data['Low'], close=data['Close'])])
figure.update_layout(title='Apple Stock Price Analysis', xaxis_rangeslider_visible=False)
figure.show()
```

Let's Check the Correlation between the features

```python
data.corr()
```

We'll use Autots library.
This library will help us for predicting the stock prices for the next 5 days.

```python
!pip install autots
```

```python
from autots import AutoTS
model = AutoTS(forecast_length=5, frequency='infer', ensemble='simple')
model = model.fit(data, date_col='Date', value_col='Close', id_col=None)
prediction = model.predict()
forecast = prediction.forecast
```
